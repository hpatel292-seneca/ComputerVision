```python
from matplotlib import pyplot as plt
import matplotlib.image as img
import numpy as np
import cv2
```

Hereâ€™s a list of the basic things you can do with **Matplotlib** in Python:

### 1. **Basic Plotting**
   - **Line Plot**: `plt.plot()`
   - **Scatter Plot**: `plt.scatter()`
   - **Bar Plot**: `plt.bar()`
   - **Horizontal Bar Plot**: `plt.barh()`
   - **Histogram**: `plt.hist()`
   - **Pie Chart**: `plt.pie()`

### 2. **Customizing Plots**
   - **Setting Titles**: `plt.title()`
   - **X and Y Axis Labels**: `plt.xlabel()`, `plt.ylabel()`
   - **Gridlines**: `plt.grid()`
   - **Legends**: `plt.legend()`
   - **Axis Limits**: `plt.xlim()`, `plt.ylim()`

### 3. **Formatting the Plot**
   - **Line Styles**: Change line types, e.g., `plt.plot(x, y, linestyle='--')` for dashed lines
   - **Marker Styles**: Use markers like circles, squares, e.g., `plt.plot(x, y, marker='o')`
   - **Colors**: Customize line and marker colors, e.g., `plt.plot(x, y, color='red')`
   - **Line Width**: Adjust line thickness, e.g., `plt.plot(x, y, linewidth=2)`
   - **Figure Size**: Set custom size, e.g., `plt.figure(figsize=(10, 5))`

### 4. **Subplots**
   - **Creating Multiple Subplots**: `plt.subplot(rows, cols, index)`
   - **Subplots with Grid**: `plt.subplots()` to create a grid of plots
   - **Adjusting Layout**: `plt.tight_layout()` to adjust subplot spacing

### 5. **Working with Data**
   - **Plotting Lists or Arrays**: Use lists or NumPy arrays for data in `plt.plot()`
   - **Handling Pandas DataFrames**: Directly plot Pandas DataFrame columns, e.g., `df.plot()`
   - **Error Bars**: Add error bars using `plt.errorbar()`

### 6. **Displaying and Saving Plots**
   - **Displaying Plot**: `plt.show()`
   - **Saving Plot as Image**: `plt.savefig('filename.png')`

### 7. **Advanced Plot Types**
   - **Stacked Bar Plot**: `plt.bar()` with stacked data
   - **Box Plot**: `plt.boxplot()`
   - **Violin Plot**: `plt.violinplot()`
   - **Heatmap**: `plt.imshow()` for displaying heatmaps
   - **Contour Plot**: `plt.contour()`, `plt.contourf()` for filled contour plots

### 8. **Customizing Axes**
   - **Logarithmic Scale**: `plt.xscale('log')`, `plt.yscale('log')`
   - **Ticks**: Custom tick marks with `plt.xticks()`, `plt.yticks()`
   - **Twin Axes**: `plt.twinx()` for two Y axes in the same plot

### 9. **3D Plotting (with `mpl_toolkits.mplot3d`)**
   - **3D Line Plot**: `ax.plot()`
   - **3D Scatter Plot**: `ax.scatter()`
   - **3D Surface Plot**: `ax.plot_surface()`

### 10. **Annotations and Text**
   - **Annotating Points**: `plt.annotate()`
   - **Adding Text**: `plt.text()` to add text at specific coordinates

### 11. **Working with Time Series Data**
   - **Plotting Time Series**: Matplotlib can plot datetime objects using `plt.plot_date()`

### 12. **Interactive Plots**
   - **Interactive Mode**: `plt.ion()` to make plots interactive
   - **Zoom and Pan**: Allows for zooming and panning in interactive mode

### 13. **Plotting Math Functions**
   - **LaTeX-style Text**: Use `r'$...$'` to render LaTeX-like mathematical expressions
   - **Plotting Functions**: Directly plot functions using `plt.plot()`

### 14. **Colormaps**
   - **Applying Colormaps**: `plt.imshow()` or `plt.scatter()` with colormap options, e.g., `cmap='viridis'`

### 15. **Annotations**
   - **Annotating Text**: `plt.text()` to add text annotations
   - **Point Annotations**: Use `plt.annotate()` to label specific points on a plot

These are the basic things you can do with **Matplotlib** to create and customize visualizations in Python.

```python
x=np.array([1,4,7,8,9])
y=np.array([5,3,8,9,2])

# We can also set title
plt.title("Bar chart")

# We can add labels
plt.xlabel("Hello")
plt.ylabel("World")
plt.plot(x,y, linestyle="--")
plt.show()
```

Shape of image is represented in (Height, width, mode) in this format, for colored image mode value is from 0 to 2 and for black and white image mode value is 0 and 1 only. In the output image, only the mode of the image is modified.

```python
modifiedImage=img1[:, :, 2]

plt.imshow(modifiedImage)
```

```python
#Modify the shape of the image
modifiedImage=img1[0:1600, 1000:2000, 1]

plt.imshow(modifiedImage)
```
