## OpenCV
- Opencv-python is a library of python bindings designed to solve computer vision problems.
- cv2.copyMakeBorder() method is used to create a border around the image like a photo frame.

Hereâ€™s how to use `cv2.copyMakeBorder()` in OpenCV to add borders to an image:

### Syntax
```python
cv2.copyMakeBorder(src, top, bottom, left, right, borderType, value)
```

### Parameters
- **`src`**: The input image (as a NumPy array).
- **`top, bottom, left, right`**: Number of pixels to add to the corresponding borders.
- **`borderType`**: Type of border to be added. It can be one of the following:
  - `cv2.BORDER_CONSTANT`: Adds a constant-colored border.
  - `cv2.BORDER_REFLECT`: Reflects the border elements (mirror reflection).
  - `cv2.BORDER_REFLECT_101` or `cv2.BORDER_DEFAULT`: Same as `BORDER_REFLECT`, but the reflection skips the border element itself.
  - `cv2.BORDER_REPLICATE`: Repeats the last pixel value on the edges.
  - `cv2.BORDER_WRAP`: The border will be wrapped around the image (circular effect).
- **`value`**: Border color, used when `borderType` is `cv2.BORDER_CONSTANT`.


````python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
import matplotlib.image as img 
````

````python
img1=img.imread('./Pizza.jpg')
constant=cv.copyMakeBorder(img1, 50, 50, 50, 50, cv.BORDER_CONSTANT)
plt.imshow(constant)
````

````python
# Lets add value to the above img
constant=cv.copyMakeBorder(img1, 50, 50, 50, 50, cv.BORDER_CONSTANT, value=[255, 0, 0])
plt.imshow(constant)
````

### Basic Uses of `randint()`:

1. **Generating Random Numbers**
   - `randint(a, b)` generates a random integer between `a` and `b` (inclusive).
   - Example:
     ```python
     from random import randint
     random_num = randint(1, 10)  # Generates a random number between 1 and 10
     ```

````python
from random import randint
constant=cv.copyMakeBorder(img1, 50, 50, 50, 50, cv.BORDER_CONSTANT, value=[randint(0,255), randint(0,255), randint(0,255)])
plt.imshow(constant)
````

````python
# cv2.BORDER_REFLECT: Reflects the border elements (mirror reflection).
constant=cv.copyMakeBorder(img1, 50, 50, 50, 50, cv.BORDER_REFLECT)
plt.imshow(constant)
````

````python
# BORDER_REPLICATE
constant=cv.copyMakeBorder(img1, 50, 50, 50, 50, cv.BORDER_REPLICATE)
plt.imshow(constant)
````

````python
# BORDER_WRAP
constant=cv.copyMakeBorder(img1, 50, 50, 50, 50, cv.BORDER_WRAP)
plt.imshow(constant)
````

The `cv2.cvtColor()` function in OpenCV is used to convert images from one color space to another. Here are the basics for using it:

### Syntax:
```python
cv2.cvtColor(src, code)
```

### Parameters:
- **`src`**: The source image (as a NumPy array).
- **`code`**: The color space conversion code, which specifies the type of conversion (e.g., BGR to Grayscale, BGR to RGB, etc.).

### Common Color Conversion Codes:
1. **BGR to Grayscale**:
   - Code: `cv2.COLOR_BGR2GRAY`
   - Example: Convert a color image to grayscale.
     ```python
     gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
     ```

2. **BGR to RGB**:
   - Code: `cv2.COLOR_BGR2RGB`
   - Example: Convert BGR (default in OpenCV) to RGB (common in image processing).
     ```python
     rgb_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
     ```

3. **BGR to HSV (Hue, Saturation, Value)**:
   - Code: `cv2.COLOR_BGR2HSV`
   - Example: Convert BGR to HSV for color detection.
     ```python
     hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
     ```

4. **RGB to Grayscale**:
   - Code: `cv2.COLOR_RGB2GRAY`
   - Example: Convert an RGB image to grayscale.
     ```python
     gray_image = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
     ```

5. **Grayscale to BGR**:
   - Code: `cv2.COLOR_GRAY2BGR`
   - Example: Convert a grayscale image back to a 3-channel BGR image.
     ```python
     bgr_image = cv2.cvtColor(gray_image, cv2.COLOR_GRAY2BGR)
     ```

````python
plt.imshow(cv.cvtColor(img1, cv.COLOR_BGR2RGB))
plt.show()
````
